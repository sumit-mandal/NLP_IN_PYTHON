```python
import re
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = 'I always uh do the main um processing, I mean, the uh um data-processing.'
```

```python
stats = nlp(doc)
```

```python
for token in stats:
    print(token.text)
```

```python
doc2 = 'U.K. has a reasonable population'
stats = nlp(doc2)

for token in stats:
    print(token.text)
```

```python
for token in re.split('\W+',doc):
    print(token)
```
