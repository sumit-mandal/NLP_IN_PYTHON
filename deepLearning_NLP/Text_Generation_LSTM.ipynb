```python
def read_file(filepath):
    with open(filepath) as f:
        str_text = f.read()
        
    return str_text
```

```python
import spacy
nlp = spacy.load('en_core_web_lg',disable=['parser','tagger','ner'])
nlp.max_length = 1198623
```

```python
def separate_punc(doc_text):
    '''Get rid of \n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n'''
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

```python
d = read_file('moby_dick_four_chapters.txt')
d[:300]
```

```python
tokens = separate_punc(d)
tokens
```

```python
len(tokens)
```

## Create sequence of tokens
###### Pass 25 words and then the network predicts 26th word

```python
train_len = 25 + 1

text_sequences = []

for i in range(train_len,len(tokens)):
    seq = tokens[i - train_len:i]
    
    text_sequences.append(seq)
```

```python
print(text_sequences[0])
print(text_sequences[1])
print(text_sequences[2])
```

```python
' '.join(text_sequences[0])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text_sequences)
```

```python
sequences = tokenizer.texts_to_sequences(text_sequences)
```

```python
import numpy as np

sequences = np.array(sequences)
sequences

# Notice that number is moving 1 time frame ahead.
```

```python
from tensorflow.keras.utils import to_categorical
X = sequences[:,:-1]
X
```

```python
y = sequences[:,-1]
y
```

```python
y = to_categorical(y,num_classes=vocabulary_size+1)
y
```

```python
seq_len = X.shape[1]
seq_len
```

```python
X.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Embedding
```

```python
def create_model(vocabulary_size,seq_len):
    
    model = Sequential()
    model.add(Embedding(vocabulary_size,seq_len,input_length = seq_len))
    model.add(LSTM(50,return_sequences=True))
    model.add(LSTM(50))
    model.add(Dense(50,activation='relu'))
    
    model.add(Dense(vocabulary_size,activation = 'softmax'))
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    
    model.summary()
    
    return model
    
```

```python
model = create_model(vocabulary_size+1,seq_len)
```

```python
from pickle import dump,load
```

```python
model.fit(X,y,batch_size=128,epochs=2,verbose=1)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences 
```

```python
def generate_text(model,tokenizer,seq_len,seed_text,num_gen_words):
    
    output_text = []
    
    input_text = seed_text
    
    for i in range(num_gen_words):
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        print(encoded_text)
        pad_encoded = pad_sequences([encoded_text],maxlen = seq_len)
        
        pred_word_ind = model.predict_classes(pad_encoded,verbose=0)[0]
        
        pred_word = tokenizer.index_word[pred_word_ind]
        
        input_text += ' '+pred_word
        
        output_text.append(pred_word)
    
    return ' '.join(output_text)
```
